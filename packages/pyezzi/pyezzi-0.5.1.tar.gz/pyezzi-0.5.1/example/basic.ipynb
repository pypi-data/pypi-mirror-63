```python
from skimage.io import imread
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np

epi = imread("epi.tif").astype(bool)
wall = imread("wall.tif").astype(bool)
s = 22

labeled_image = np.zeros_like(epi, np.uint8)

labeled_image[epi] = 1
labeled_image[wall] = 2

f, ax = plt.subplots()
plt.imshow(labeled_image[s])
plt.colorbar()
plt.show()
```

```python
from pyezzi import compute_thickness
thickness = compute_thickness(labeled_image)

f, ax = plt.subplots()
plt.imshow(thickness[s])
plt.colorbar()
plt.show()
```

```python
from skimage.io import imsave
imsave('/tmp/thickness.tif', thickness)
```
